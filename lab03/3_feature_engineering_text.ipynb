```python
from newmm_tokenizer.tokenizer import word_tokenize

text = data["text"].iloc[0]
text, repr(word_tokenize(text))
```

## N-gram vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

# vectorizer = CountVectorizer()
vectorizer = CountVectorizer(tokenizer=word_tokenize)
# vectorizer = CountVectorizer(tokenizer=word_tokenize, ngram_range=(1, 2))
vectorizer.fit(data["text"])
len(vectorizer.vocabulary_)
```

```python
repr(vectorizer.vocabulary_)[:100]
```

```python
vocabulary = vectorizer.vocabulary_.items()
vocabulary = sorted(vocabulary, key=lambda token_index: token_index[1])
vocabulary = [token for token, index in vocabulary]
print(repr(vocabulary[:10]))
```

```python
vector = vectorizer.transform([text]).A[0]
text, repr(word_tokenize(text)), repr(vocabulary[:100]), repr(list(vector))[:100]
```

```python
counts = pd.DataFrame([vector], columns=vocabulary)
counts
```
