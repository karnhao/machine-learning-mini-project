```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("house_prices.csv",encoding="unicode_escape")
data.rename(columns = {"Price (in rupees)": "Price"}, inplace = True)
data
```

```python
data.drop(inplace = True, labels = ["Dimensions", "Plot Area", "facing", "Description", "Society", "Title", "Floor", "Status", "Super Area"], axis = 1)
data.drop([], axis = 1, inplace = True)
data
```

```python
data.drop(inplace = True, labels = "Index", axis=1)
data
```

```python
data.drop_duplicates(inplace = True)
data
```

```python
data["Car Parking"].replace(np.nan, 0, inplace=True)
data
```

```python
data.dropna(inplace = True)
data
```

```python
data["Carpet Area"] = data["Carpet Area"].map(lambda x: int(x.split(" ")[0]))
data
```

```python
open_covered = data["Car Parking"].map(lambda x: "None" if x == 0 else x.replace(",", "")).map(lambda x : x if x == "None" else x.split(" ")[1])
data["open_covered"] = open_covered
data
```

```python
car_parking_count = data["Car Parking"]
car_parking_count = car_parking_count.map(lambda x: x if isinstance(x, int) else int(x.split(" ")[0]))
data["car_parking_count"] = car_parking_count
data
```

```python
data.drop(labels="Car Parking", axis=1, inplace=True)
data
```

```python
data.drop("Amount(in rupees)", axis=1, inplace=True) # Data Leak
data
```

```python
# Remove rows with all zeros in column 'Price'
data = data.loc[~(data['Price'] == 0)]
```

```python
data.reset_index(inplace = True, drop=True)
data
```

```python
data["Bathroom"].replace("> 10", 11, inplace = True)  # แก้ไขข้อมูล ">10" ให้เป็น 11 
data["Bathroom"].unique()
```

```python
data["Balcony"].replace("> 10", 11, inplace=True)
data["Balcony"].unique()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.3, random_state=0)
valid_data, test_data = train_test_split(test_data, test_size=0.5, random_state=0)
train_data.shape[0], valid_data.shape[0], test_data.shape[0]
```

```python
label = "Price"
train_data[label].describe()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler

label_pipeline = make_pipeline(
    FunctionTransformer(np.log2, inverse_func=np.exp2),
    StandardScaler(),
)

train_data[label] = label_pipeline.fit_transform(train_data[[label]])
valid_data[label] = label_pipeline.transform(valid_data[[label]])
test_data[label] = label_pipeline.transform(test_data[[label]])

train_data[label]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer as TfidfVec
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor

def create_pipeline():
    return make_pipeline(
        make_column_transformer(
            (OneHotEncoder(handle_unknown="ignore"), ["Transaction"]),
            (OneHotEncoder(handle_unknown="ignore"), ["Furnishing"]),
            (OneHotEncoder(handle_unknown="ignore"), ["Ownership"]),
            (OneHotEncoder(handle_unknown="ignore"), ["open_covered"]),
            (OneHotEncoder(handle_unknown="ignore"), ["overlooking"]),
            (OneHotEncoder(handle_unknown="ignore"), ["location"]),
            (StandardScaler(), ["Bathroom"]),
            (StandardScaler(), ["Balcony"]),
            (StandardScaler(), ["Carpet Area"]),
            (StandardScaler(), ["car_parking_count"]),
            # (TfidfVec(ngram_range=(1, 20), max_df = len(locations), min_df = 1), ["location"]),
            remainder="passthrough"
        ),
        DecisionTreeRegressor(random_state=0),
    )

pipeline = create_pipeline()
pipeline
```

```python
pipeline.fit(train_data[features], train_data[label])

train_score = pipeline.score(train_data[features], train_data[label])
print(f"{train_score:,.3f}")
```

```python
from sklearn.model_selection import GridSearchCV

params = dict(
   decisiontreeregressor__criterion = ['friedman_mse'],
   decisiontreeregressor__max_depth = [16],
   decisiontreeregressor__max_features = [None]
)

search = GridSearchCV(create_pipeline(), params, cv=5, n_jobs=-1)
search.fit(train_data[features], train_data[label])

print(f"{search.best_score_:,.3f}", search.best_params_)
```

```python
test_score = search.best_estimator_.score(test_data[features], test_data[label])
print(f"{test_score:,.3f}")
```

```python
%%writefile app.py
import pandas as pd
import gradio as gr
import joblib
# "Artifacts"
pipeline = joblib.load("pipeline.joblib")
label_pipeline = joblib.load("label_pipeline.joblib")

ownerships = joblib.load("ownerships.joblib")
transactions = joblib.load("transactions.joblib")
locations = joblib.load("locations.joblib")
furnishings = joblib.load("furnishings.joblib")
open_covereds = joblib.load("open_covereds.joblib")
overlooking = joblib.load("overlooking.joblib")

def predict(ownerships, transactions, locations, furnishings, open_covereds,carpet_area,overlooking,bathroom,balcony,car_parking):
    sample = dict()
    sample["Carpet Area"] = carpet_area
    sample["overlooking"] = overlooking
    
    sample["Bathroom"] = bathroom # Column names matching feature names
    sample["Balcony"] = balcony
    sample["car_parking_count"] = car_parking
    
    sample["Ownership"] = ownerships
    sample["Transaction"] = transactions
    sample["location"] = locations
    sample["Furnishing"] = furnishings
    sample["open_covered"] = open_covereds

    price = pipeline.predict(pd.DataFrame([sample]))
    price = label_pipeline.inverse_transform([price])
    
    return int(price[0][0])

# https://www.gradio.app/guides
with gr.Blocks() as blocks:
   
    location = gr.Dropdown(locations, value=locations[0], label="location")
    carpet_area = gr.Number(label="Carpet Area", value=1, minimum=1)
    bathroom = gr.Slider(label="Bathroom", minimum=0, maximum=10, step=1,info = "How many bathrooms are there?")
    balcony = gr.Slider(label="Balcony", minimum=0, maximum=10, step=1,info = "How many balcony are there?")
    car_parking = gr.Slider(label="car_parking_count", minimum=0, maximum=10, step=1,info = "How many car parking are there?")
    transactions = gr.Radio(transactions, label="Transaction")            
    overlooking = gr.CheckboxGroup(overlooking, label="overlooking")            
    ownerships = gr.Radio(ownerships, label="Ownership") 
    furnishings = gr.Radio(furnishings, label="Furnishing") 
    open_covereds = gr.Radio(open_covereds, label="open_covered") 



    price = gr.Number(label="Price")

    inputs = [ownerships, transactions, locations, furnishings, open_covereds,carpet_area,overlooking,bathroom,balcony,car_parking]
    outputs = [price]

    predict_btn = gr.Button("Predict")
    predict_btn.click(predict, inputs=inputs, outputs=outputs)

if __name__ == "__main__":
    blocks.launch() # Local machine only
    # blocks.launch(server_name="0.0.0.0") # LAN access to local machine
    # blocks.launch(share=True) # Public access to local machine
```
