```python
from sklearn.feature_extraction.text import HashingVectorizer

for feature in ["Ownership", "overlooking"] :
    hasher = HashingVectorizer(n_features=4, binary=True)

    encoded = hasher.fit_transform(data[feature])
    encoded = pd.DataFrame(encoded.A, columns=[f"{feature}_{i}" for i in range(hasher.n_features)])
    encoded

    data = pd.concat([data, encoded.set_index(data.index)], axis=1)



data
```
