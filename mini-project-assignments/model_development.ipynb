```python
import pandas as pd
```

```python
data = pd.read_csv("./house_prices_fix.csv")
data
```

```python
data.drop("Floor", axis=1, inplace=True)
data
```

```python
data["overlooking_1"].unique()
```

```python
data.drop("overlooking_1", axis=1, inplace=True)
```

```python
data["Bathroom"].unique()
```

```python
data["Bathroom"].replace("> 10", 11, inplace=True)
data["Bathroom"].unique()
```

```python
data["Balcony"].unique()
```

```python
data["Balcony"].replace("> 10", 11, inplace=True)
data["Balcony"].unique()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.3, random_state=0)
test_data, valid_data = train_test_split(test_data, test_size=0.5, random_state=0)
train_data.shape[0], valid_data.shape[0], test_data.shape[0]
```

```python
features = data.columns[1:].tolist()

train_data[features]
```

```python
label = "Price (in rupees)"
train_data[label]
```

```python
train_data[label].describe()
```

```python

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler

import numpy as np

label_tfm = make_pipeline(
    FunctionTransformer(
        func=lambda x: np.log10(x + 1),
        inverse_func=lambda x: 10 ** x,
    ),
    StandardScaler(),
)

label_tfm.fit_transform(train_data[label].to_frame())
```

```python
train_data[label] = label_tfm.transform(train_data[label].to_frame())
valid_data[label] = label_tfm.transform(valid_data[label].to_frame())
test_data[label] = label_tfm.transform(test_data[label].to_frame())

train_data[label]
```

```python
train_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score

baseline = LinearRegression(n_jobs=-1)
baseline.fit(train_data[features], train_data[label])

score = baseline.score(test_data[features], test_data[label])
print(f"Testing Score: {score:.3f}")

score = baseline.score(train_data[features], train_data[label])
print(f"Training Score: {score:.3f}")


kfold = KFold(shuffle=False, n_splits=10)
score = cross_val_score(baseline, train_data[features], train_data[label], cv=kfold)
print(f"Cross-Validation Score Mean: {score.mean():.3f}")
```

```python
from sklearn.tree import DecisionTreeRegressor


model = DecisionTreeRegressor()
model.fit(train_data[features], train_data[label])

score = model.score(test_data[features], test_data[label])
print(f"Testing Score: {score:.3f}")

score = model.score(train_data[features], train_data[label])
print(f"Training Score: {score:.3f}")

score = cross_val_score(model, train_data[features], train_data[label], cv=kfold)
print(f"Cross-Validation Score Mean: {score.mean():.3f}")
```

```python
from sklearn.svm import SVR


model = SVR()
model.fit(train_data[features], train_data[label])

score = model.score(test_data[features], test_data[label])
print(f"Testing Score: {score:.3f}")

score = model.score(train_data[features], train_data[label])
print(f"Training Score: {score:.3f}")

score = cross_val_score(model, train_data[features], train_data[label], cv=kfold)
print(f"Cross-Validation Score Mean: {score.mean():.3f}")

```

```python
from sklearn.ensemble import BaggingRegressor

model = BaggingRegressor(random_state=0)
model.fit(train_data[features], train_data[label])

score = model.score(test_data[features], test_data[label])
print(f"Testing Score: {score:.3f}")

score = model.score(train_data[features], train_data[label])
print(f"Training Score: {score:.3f}")

score = cross_val_score(model, train_data[features], train_data[label], cv=kfold)
print(f"Cross-Validation Score Mean: {score.mean():.3f}")
```

```python
from sklearn.ensemble import AdaBoostRegressor

model = AdaBoostRegressor(random_state=0)
model.fit(train_data[features], train_data[label])

score = model.score(test_data[features], test_data[label])
print(f"Testing Score: {score:.3f}")

score = model.score(train_data[features], train_data[label])
print(f"Training Score: {score:.3f}")

score = cross_val_score(model, train_data[features], train_data[label], cv=kfold)
print(f"Cross-Validation Score Mean: {score.mean():.3f}")
```
